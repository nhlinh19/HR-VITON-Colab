```python
import os
import re
import subprocess

def test_generator(checkpoint):
    # Get the path to the checkpoint folder
    checkpoint_path = f"./checkpoints/{checkpoint}"
    
    # Find the latest gen_step_x.pth file in the folder
    latest_gen_file = None
    latest_step = -1
    for filename in os.listdir(checkpoint_path):
        match = re.match(r"gen_step_(\d+)\.pth", filename)
        if match:
            step = int(match.group(1))
            if step > latest_step:
                latest_step = step
                latest_gen_file = filename
    
    if latest_gen_file is None:
        print("No gen_step_x.pth files found in checkpoint folder.")
        return
    
    # Build the paths to the gen_step_x.pth file and output directory
    gen_file_path = os.path.join(checkpoint_path, latest_gen_file)
    output_dir_path = f"./output/{checkpoint}"

    print(gen_file_path)
    print(output_dir_path)
    
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir_path, exist_ok=True)
    
    # Run the test_generator.py script with the appropriate arguments
    # cmd = f"python test_generator.py --gen_checkpoint={gen_file_path} --output_dir={output_dir_path}"
    !python test_generator.py --gen_checkpoint="$gen_file_path" --output_dir="$output_dir_path"
    # subprocess.run(cmd, shell=True)

checkpoints = ['expHingeFeat-512-30', 'expHingeL1-512-30']
for checkpoint in checkpoints:
    test_generator(checkpoint)
```

```python
import os

def delete_old_files(folder_path: str):
    dis_files = []
    gen_files = []
    for file in os.listdir(folder_path):
        if file.startswith('dis_step_') and file.endswith('.pth'):
            dis_files.append(int(file.split('_')[2].split('.')[0]))
        elif file.startswith('gen_step_') and file.endswith('.pth'):
            gen_files.append(int(file.split('_')[2].split('.')[0]))
    
    max_dis_file = 'dis_step_' + str(max(dis_files)).zfill(6) + '.pth'
    max_gen_file = 'gen_step_' + str(max(gen_files)).zfill(6) + '.pth'

    print(max_dis_file)
    print(max_gen_file)
    
    for file in os.listdir(folder_path):
        if (file.startswith('dis_step_') or file.startswith('gen_step_')) and file.endswith('.pth'):
            if file != max_dis_file and file != max_gen_file:
                # print(file)
                os.remove(os.path.join(folder_path, file))

# Example usage
delete_old_files('/root/virtual-try-on/HR-VITON-Colab/checkpoints/expHingeL1-512-20v2')
```

```python
import os

def print_eval_files(experiment_folders):

    for experiment_folder in experiment_folders:
        eval_file_path = os.path.join('output', experiment_folder, 'eval.txt')
        if os.path.isfile(eval_file_path):
            with open(eval_file_path, 'r') as eval_file:
                print(f'Experiment: {experiment_folder}')
                print(eval_file.read())
                # print('---')
        else:
            print(f'No eval.txt file found in {experiment_folder}')
            print('---')

# Example usage
# experiment_folders = ['expHinge-512', 'expHingeL1-512', 'expLs-512', 'expLsL1-512', 'expOriginal-512', 'expOriginalL1-512']
experiment_folders = ['expHingeFeat-512-30', 'expHingeL1-512-30']
print_eval_files(experiment_folders)
```
